# State problem

Fokker-Planck equation
\begin{cases}
      \dfrac{\partial y}{\partial t} + \nabla \cdot (- \nu \nabla y + \mathbf{u} y) = 0 \qquad &\mathrm{in} \ \Omega \times (0,T)  \\
             (-\nu \nabla y + \mathbf{u} y) \cdot \mathbf{n} = 0  \qquad &\mathrm{on} \ \partial \Omega  \times (0,T) \\
       y(0) = \bar{y} \qquad &\mathrm{in} \ \Omega \times \{t = 0\}
\end{cases}

where $\Omega$ is the domain of interest (visualised below) with boundary $\partial \Omega$, $T>0$ is the final time, $y: \Omega \times [0,T] \to \mathbb R$ is the unknown state field, $\mathbf{u}:\Omega times [0,T) \to \mathbb R^2$ is the control velocity field, $\nu$ is the diffusion coefficient, $\mathbf{n}$ is the normal versor and $\bar{y}$ is the initial datum

```python
# CREATE THE MESH FOR THE STATE PROBLEM

domain = geo.Rectangle((-1.0, -1.0), (1.0, 1.0))
mesh = geo.mesh(domain, stepsize = 0.05)
mesh = refine(mesh)

plot(mesh, color = "grey", linewidth = 0.75)
plt.title("Mesh ($N_{h}$=%d)" % mesh.num_vertices())
plt.axis("off");
```

```python
# DEFINE MEASURES

dx = Measure("dx", domain = mesh)

ds = Measure("ds", domain = mesh)
```

```python
# DEFINE THE CONTROL AND THE STATE SPACE

Uh = VectorElement('CG', mesh.ufl_cell(), 1)
Uh = FunctionSpace(mesh, Uh)
ncontrol = Uh.dim()

Yh = Uh.sub(0).collapse()
nstate = Yh.dim()
```

```python
# DEFINE THE STATE PROBLEM SOLVER

diffusion = 0.001
beta = 0.1
beta_g = 0.1

def gaussian_generator(params):
    """
    Generate a 2D Gaussian starting from the mean position
    Input: mean coordinate
    Output: 2D Gaussian function
    """
    
    y0 = Expression('10 / pi * exp(- 10*(x[0] - x0)*(x[0] - x0) - 10*(x[1] - x1)*(x[1] - x1))', degree = 1, x0 = params[0], x1 = params[1])
    return interpolate(y0, Yh)
    
def compute_state(ut, y0, dt, ntimesteps, yd = None):
    """
    Compute the state trajectory
    Input: control, initial condition, time step, number of timesteps and target state
    Output: time-dendent solution of the Fokker-Planck equation
    """
    
    u = Function(Uh)

    yt = CPU.zeros(ntimesteps + 1, nstate)
    yt[0] = CPU.tensor(y0.vector())

    J = 0.0

    i = 0

    for i in range(ntimesteps):

        u.assign(ut[i])
            
        w = TestFunction(Yh)
        y = TrialFunction(Yh)

        a = inner(y, w) * dx + 0.5 * dt * diffusion * inner(grad(y), grad(w)) * dx - 0.5 * dt * y * project(u.sub(0), Yh) * w.dx(0) * dx - 0.5 * dt * y * project(u.sub(1), Yh) * w.dx(1) * dx
        L = inner(y0, w) * dx - 0.5 * dt * diffusion * inner(grad(y0), grad(w)) * dx + 0.5 * dt * y0 * project(u.sub(0), Yh) * w.dx(0) * dx + 0.5 * dt * y0 * project(u.sub(1), Yh) * w.dx(1) * dx
    
        y = Function(Yh)
        solve(a == L, y)
        
        y0.assign(y)
        yt[i+1] = CPU.tensor(y.vector())

        if yd is not None:
            J += 0.5 * assemble(inner(y - yd, y - yd) * dx + inner(y, y) * ds + beta * inner(u, u) * dx + beta * inner(grad(u), grad(u)) * dx)
             
    return yt, J
```

```python
# SOLVE THE STATE PROBLEM SOLVER

dt = Constant(0.25)
T = 1.0
ntimesteps = round(T / dt)

y0 = gaussian_generator([-0.5, 0.0])
yd = gaussian_generator([0.5, 0.0])

ut = []
for i in range(ntimesteps):
    ut.append(Function(Uh))
    
yt, J = compute_state(ut, y0, dt, ntimesteps, yd)
clc()

set_working_tape(Tape())    

fe.plot(yt[-1], Yh, levelcurves = True, cmap = "jet")
plt.axis('off');
```

# Snapshots generation

### PDE-constrained Optimisation

Consider the optimal transport problem

$$
\min_{y,\mathbf{u}} \frac{1}{2} \int_0^T \int_{\Omega} (y - y_d)^2 + \int_0^T \int_{\partial \Omega} y^2 + \frac{\beta}{2} \int_0^T \int_{\Omega} ||\mathbf{u}||^2 + \frac{\beta_g}{2}  \int_0^T \int_{\Omega} ||\nabla \mathbf{u}||^2 
$$

where $y_d$ is the target state. Physically, this setup corresponds to minimising the discrepancy between current and target state over the domain $\Omega$ and in the time interval $(0,T]$, keeping the state far away from the domain boundaries $\partial \Omega$. To avoid overconsuming and irregular controls, the norm of the control and its gradient are penalised via the regularisation terms.

```python
# GENERATE SNAPSHOTS (SKIP THIS CELL IF DATA ALREADY AVAILABLE)

from dlroms.ocp import snapshots

dt = Constant(0.25)
T = 1.0
ntimesteps = round(T / dt)

ntrajectories = 100
ntrain = round(0.8 * ntrajectories)
ntrainY = ntrain * (ntimesteps + 1)
ntrain *= ntimesteps
nsnapshots = ntrajectories * ntimesteps
nparams = 2

y0_ranges = [(-0.5, 0.0), (-0.5, 0.5)] # Initial datum parameters: mean coordinates of the initial datum
yd_ranges = [(0.0, 0.5), (-0.5, 0.5)] # Scenario paramters: mean coordinates of the target position

set_working_tape(Tape()) # Reset forward model annotated by dolfin adjoint 

def sampler(seed):
    """
    Generate a random sample of scenario parameters and the corresponding optimal pair
    Input: seed
    Output: scenario parameters, optimal state and control
    """
    
    y0_params = [0] * len(y0_ranges)
    for i in range(len(y0_ranges)):
            y0_params[i] = ((y0_ranges[i][1] - y0_ranges[i][0]) * CPU.rand() + y0_ranges[i][0]).item()  
    y0 = gaussian_generator(y0_params)

    yd_params = [0] * len(yd_ranges)
    for i in range(len(yd_ranges)):
            yd_params[i] = ((yd_ranges[i][1] - yd_ranges[i][0]) * CPU.rand() + yd_ranges[i][0]).item()  
    yd = gaussian_generator(yd_params)

    ut = []
    i = 0
    for i in range(ntimesteps):
        ut.append(Function(Uh))
    
    yt, J = compute_state(ut, y0, dt, ntimesteps, yd)

    control = [Control(u) for u in ut]

    Jhat = ReducedFunctional(J, control)

    ut_opt = minimize(Jhat, method = 'L-BFGS-B', options = {'disp': True, 'maxiter': 500})

    y0_new = gaussian_generator(y0_params) # New y0 since the original y0 modified in compute_state
    
    yt_opt, J_opt = compute_state(ut_opt, y0_new, dt, ntimesteps, yd)

    set_working_tape(Tape())    

    ut_opt_vals = CPU.zeros(ntimesteps, ncontrol)
    yd_params_vals = CPU.zeros(ntimesteps, nparams)
    i = 0
    for i in range(ntimesteps):
        ut_opt_vals[i] = CPU.tensor(ut_opt[i].vector())
        yd_params_vals[i] = CPU.tensor(yd_params)
        i += 1

    return yd_params_vals, yt_opt, ut_opt_vals

snapshots(ntrajectories, sampler, core = CPU, verbose = True, filename = 'Data/OptimalTransport')
```

```python
# MERGE FUNCTION TO BUILD THE SNAPSHOTS MATRIX FROM ITS X AND Y COMPONENTS

def merge_U(Ux, Uy):
    ncontrol = Ux.shape[1] + Uy.shape[1]
    U = CPU.zeros(Ux.shape[0], ncontrol)
    U[:, 0 : ncontrol : 2] = Ux
    U[:, 1 : ncontrol : 2] = Uy
    return U
```

```python
# DEFINE A FUNCTION TO PLOT THE DISTANCE BETWEEN STATE AND TARGET IN TIME

def plot_J(Y, MU):
    """
    Plot the distance between state and target in time
    Input: state trajectory, scenario parameters
    """
    yd = gaussian_generator(MU)
    vals = []
    times = [i*float(dt) for i in range(0, Y.shape[0])]
    
    for i in range(0, Y.shape[0]):
        yt = fe.asvector(Y[i], Yh)
        vals.append(dolfin.assemble(sqrt((yt - yd)**2) * dx))
        
    plt.plot(times, [vals[0]]*len(times), label = "$\mathbf{u}_h = \mathbf{0}$", color = teal[14], linewidth = 3)
    plt.plot(times, vals, label = "$\mathbf{u}_h = \mathbf{u}^*_h$", color = magenta[5], linewidth = 3)
   
    plt.plot(times, CPU.zeros(len(vals)), '--', linewidth = 1, color = "darkgray")
    plt.legend(loc = "best", fontsize=12)
    plt.xlabel(r"$t$", fontsize=15)
    plt.ylabel(r"$||\mathbf{y}_h(t) - \mathbf{y}_d(t)||_{L^2(\Omega)}$", fontsize=15)
```

## Latent loop

Forward model: $\hspace{0.1cm} \mathbf{y}_{N}(t_{j+1}) = \varphi_N (\mathbf{y}_{N}(t_j), \mathbf{u}_N(t_j), \boldsymbol{\mu}_{s}) \quad \forall j = 1,...,N_t-1$

where $ \mathbf{y}_{N} = \varphi_E^y(\mathbf{y}_{POD}) \;$ and $\; \mathbf{u}_{N} = \varphi_E^u(\mathbf{u}_{POD}) \;$

```python
# AUTOENCODERS, POLICY AND FORWARD SURROGATE MODEL DEFINITIONS

ny = 10 # State latent dimension
encoder_Y = Dense(ky, 100) + Dense(100, ny)
decoder_Y = Dense(ny, 100) + Dense(100, 100) + Dense(100, ky, activation = None)
autoencoder_Y = encoder_Y + decoder_Y

nu = 18 # Latent dimension
encoder_U = Dense(kux + kuy, 100) + Dense(100, nu)
decoder_U = Dense(nu, 200) + Dense(200, 200) + Dense(200, kux + kuy, activation = None)
autoencoder_U = encoder_U + decoder_U

policy = Dense(ny + nparams, 50) + Dense(50, 50) + Dense(50, 50) + Dense(50, nu, activation = None)

phi = Dense(nparams + ny + nu, 50) + Dense(50, 50) + Dense(50, 50) + Dense(50, ny, activation = None)
phi.He() # Initialization
```

```python
# TRANSFER LEARNING

autoencoder_Y.load_state_dict(torch.load('NN/OptimalTransportVacuum/autoencoder_Y'))
autoencoder_U.load_state_dict(torch.load('NN/OptimalTransportVacuum/autoencoder_U'))
policy.load_state_dict(torch.load('NN/OptimalTransportVacuum/policy'))

autoencoder_Y.unfreeze()
autoencoder_U.unfreeze()
policy.unfreeze()
```

```python
#  NEURAL NETWORKS TRAINING

Y_POD_reconstructed, U_POD_reconstructed, U_POD_hat, Y1_POD_hat = ocp.latent_loop(ntrajectories, Y_POD, U_POD, MU, encoder_Y, decoder_Y, encoder_U, decoder_U, policy, phi, training = False, initialization = False, save = False, path = 'NN/OptimalTransportVacuum/Latentloop/', epochs = 500, weights = [0.001, 0.001, 0.001, 1.0, 1.0, 0.001], optim = torch.optim.LBFGS, loss = mse, error = mre, best = False)
```
